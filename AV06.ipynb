```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (1 - np.cos(x))/x**2

valores_x = np.logspace(-1, -15, num=100, base=10)  # vetor com 15 valores de x entre 10^-1 e 10^-15

valores_f = f(valores_x)  # vetor com os valores de f(x) para cada x

'''print(valores_x)
print(valores_f)'''

plt.scatter(valores_x, valores_f, color='blue', label='f(x) = (1 - cos(x)) / x^2', s=10)
plt.xscale('log')
plt.title('Gráfico da função $f(x) = \\frac{1 - \\cos(x)}{x^2}$')
plt.xlabel('x (escala logarítmica)')
plt.ylabel('f(x)')
```

Item (B) 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (2*(np.sin(x/2))**2)/x**2

def g(x):
    return (1 - np.cos(x))/x**2

valores_x = np.logspace(-1, -15, num=100, base=10)  # vetor com 15 valores de x entre 10^-1 e 10^-15

valores_f = f(valores_x)  # vetor com os valores de f(x) para cada x

valores_g = g(valores_x)  # vetor com os valores de g(x) para cada x

'''print(valores_x)
print(valores_f)'''

plt.scatter(valores_x, valores_f, color='blue', label='$f(x) = \\frac{2 \\sin^2(\\frac{x}{2})}{x^2}$', s=10)
plt.scatter(valores_x, valores_g, color='red', label='$g(x) = \\frac{1 - \\cos(x)}{x^2}$', s=10)
plt.xscale('log')
plt.legend()
plt.title('Gráfico das funções $f(x) = \\frac{2 \\sin^2(x/2)}{x^2}$ e $g(x) = \\frac{1 - \\cos(x)}{x^2}$')
plt.xlabel('x (escala logarítmica)')
plt.ylabel('f(x) e g(x)')


# A função f(x) apresenta uma melhora significativa em relação à função g(x) para valores de x muito próximos de zero, pois g(x) sofre cancelamento numérico devido à subtração de dois números muito próximos, enquanto f(x) não sofre esse cancelamento.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (1 - np.cos(x))/x**2

valores_x = np.logspace(-1, -15, num=100, base=10)

valores_f_32 = f(np.float32(valores_x))  # precisão 32 bits
valores_f_64 = f(np.float64(valores_x))  # precisão 64 bits

valor_esperado = 0.5  # limite teórico para x -> 0

# Erro relativo: |valor_calculado - valor_esperado| / |valor_esperado|
erro_rel_32 = np.abs(valores_f_32 - valor_esperado) / np.abs(valor_esperado)
erro_rel_64 = np.abs(valores_f_64 - valor_esperado) / np.abs(valor_esperado)

plt.plot(valores_x, erro_rel_32, label='Erro relativo 32 bits', color='blue')
plt.plot(valores_x, erro_rel_64, label='Erro relativo 64 bits', color='red')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('x (escala logarítmica)')
plt.ylabel('Erro relativo')
plt.title('Erro relativo de $f(x)$ em diferentes precisões')
plt.legend()
plt.show()
```

Observando o gráfico que relaciona o erro relativo de cada uma das precisões, o erro relativo para float64 permanece menor e mais estável para valores pequenos de x, enquanto o erro relativo para float32 cresce rapidamente devido à menor precisão e maior suscetibilidade a erros de arredondamento e cancelamento numérico.

A precisão de 32 bits utiliza menos memória e realiza os cálculos mais rapidamente, porém, apresenta um maior erro numérico.
A precisão de 64 bits utiliza mais memória e leva um pouco mais de tempo para realizar os cálculos, porém, oferece maior precisão e estabilidade numérica, reduzindo erros de arredondamento e cancelamento numérico.

```python
import numpy as np
import matplotlib.pyplot as plt
from decimal import Decimal, getcontext

n = 10**6
valor_esperado = np.pi**2 / 6  # limite teórico da série harmônica

soma = 0.0
for i in range(1, n + 1):
    soma += 1 / i

print("Soma da série: ", soma) 
print("Valor esperado: ", valor_esperado)

print("Erro absoluto: ", np.abs(soma - valor_esperado))
print("Erro relativo: ", np.abs(soma - valor_esperado) / np.abs(valor_esperado))


# O resultado obtido para a série harmônica diverge do valor esperado, o que indica que a série não converge para o valor teórico. Isso ocorre devido à natureza da série harmônica, que diverge lentamente.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Série harmônica com i crescendo de 1 até n

n = 10**6
valor_esperado = np.pi**2 / 6  # limite teórico da série harmônica

soma = 0.0
for i in range(1, n + 1):
    soma += 1 / i

print("Soma da série: ", soma) 
print("Valor esperado: ", valor_esperado)

print("Erro absoluto: ", np.abs(soma - valor_esperado))
print("Erro relativo: ", np.abs(soma - valor_esperado) / np.abs(valor_esperado))


# Série harmônica com i decrescendo de n até 1

soma_decrescente = 0.0

for i in range(n, 0, -1):
    soma_decrescente += 1 / i

print("Soma da série decrescente: ", soma_decrescente)
print("Valor esperado: ", valor_esperado)
print("Erro absoluto (crescente): ", np.abs(soma - valor_esperado))
print("Erro relativo (crescente): ", np.abs(soma - valor_esperado) / np.abs(valor_esperado))
print("Erro absoluto (decrescente): ", np.abs(soma_decrescente - valor_esperado))
print("Erro relativo (decrescente): ", np.abs(soma_decrescente - valor_esperado) / np.abs(valor_esperado))
```

```python
# Série harmônica com i crescendo de 1 até n

n = 10**8
valor_esperado = np.pi**2 / 6  # limite teórico da série harmônica

soma = 0.0
for i in range(1, n + 1):
    soma += 1 / i

# Série harmônica com método de Kahan para evitar cancelamento numérico
soma_kahan = 0.0
c = 0.0
for i in range(1, n + 1):
    y = 1 / i - c  # compensação
    t = soma_kahan + y  # soma com compensação
    c = (t - soma_kahan) - y  # nova compensação
    soma_kahan = t  # atualiza a soma


print("Soma: ", soma)
print("Soma com Kahan:", soma_kahan)
print("Valor esperado:", valor_esperado)
print("Erro absoluto:", np.abs(soma - valor_esperado))
print("Erro relativo:", np.abs(soma - valor_esperado) / np.abs(valor_esperado))
print("Erro absoluto (Kahan):", np.abs(soma_kahan - valor_esperado))
print("Erro relativo (Kahan):", np.abs(soma_kahan - valor_esperado) / np.abs(valor_esperado))
```
