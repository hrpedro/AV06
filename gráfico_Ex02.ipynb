```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do problema
n = 10000
dt = 2 * np.pi / n
x = np.zeros(n + 1)
v = np.zeros(n + 1)
x[0] = 0
v[0] = 1

# Integração pelo método de Euler
for i in range(n):
    x[i + 1] = x[i] + v[i] * dt
    v[i + 1] = v[i] - x[i] * dt

# Vetor de tempo
t = np.linspace(0, 2 * np.pi, n + 1)

# Gráfico conjunto de posição e velocidade
plt.figure(figsize=(10, 5))
plt.plot(t, x, label='Posição x(t)')
plt.plot(t, v, label='Velocidade v(t)')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Posição e Velocidade da partícula em função do tempo')
plt.legend()
plt.grid(True)
plt.show()
```
